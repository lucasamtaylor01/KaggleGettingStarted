# Data Cleaning

```python
from pathlib import Path
import pandas as pd
import numpy as np
```

```python
# PATH DEFINITIONS
BASE_DIR = Path().resolve().parent

DATA_DIR = BASE_DIR / "data"
RAW_DATA_DIR = DATA_DIR / "data_raw"

CLEAN_DATA_DIR = DATA_DIR / "data_clean"
CLEAN_DATA_DIR.mkdir(exist_ok=True)
```

```python
df_raw = df_raw = pd.read_csv(RAW_DATA_DIR / "train.csv")
```

## String treatment

```python
df_str_treat = df_raw.copy()
```

```python
df_str_treat.columns = df_str_treat.columns.str.upper()
columns_str = df_str_treat.select_dtypes(include='string')

for col in columns_str.columns:
    df_str_treat[col] = df_str_treat[col].str.upper()

pd.set_option('display.max_columns', None)
df_str_treat.head()
```

## Null treatment

```python
df_null_treat = df_str_treat.copy()
```

```python
# GARAGE ANALYSIS

garage_columns = ['GARAGEQUAL', 'GARAGEFINISH', 'GARAGETYPE', 'GARAGEYRBLT','GARAGECOND']

df_null_treat_garage = df_null_treat[garage_columns].copy()
df_null_treat_garage = df_null_treat_garage[df_null_treat_garage['GARAGEYRBLT'].isnull()].reset_index(drop='True')
df_null_treat_garage.isnull().sum()
```

```python
# GARAGEYRBLT 
df_null_treat['HASGARAGE'] = np.where(df_null_treat['GARAGEYRBLT'].isnull(), 0, 1).astype('int8')
df_null_treat = df_null_treat.drop(columns='GARAGEYRBLT')

garage_columns = ['GARAGEQUAL', 'GARAGEFINISH', 'GARAGETYPE', 'GARAGECOND', 'HASGARAGE']

# OTHERS
df_null_treat['GARAGEQUAL'] = df_null_treat['GARAGETYPE'].fillna('NA')
df_null_treat['GARAGEFINISH'] = df_null_treat['GARAGETYPE'].fillna('NA')
df_null_treat['GARAGETYPE'] = df_null_treat['GARAGETYPE'].fillna('NA')
df_null_treat['GARAGECOND'] = df_null_treat['GARAGETYPE'].fillna('NA')

df_null_treat[garage_columns].isnull().sum()
```

```python
bsmt_columns = ['BSMTFINTYPE2', 'BSMTEXPOSURE', 'BSMTCOND', 'BSMTQUAL', 'BSMTFINTYPE1']

df_null_treat_bsmt = df_null_treat[bsmt_columns].copy()
df_null_treat_bsmt.isnull().sum()
```

```python
# OTHERS
df_null_treat['BSMTCOND'] = df_null_treat['BSMTCOND'].fillna('NA')
df_null_treat['BSMTQUAL'] = df_null_treat['BSMTQUAL'].fillna('NA')
df_null_treat['BSMTFINTYPE1'] = df_null_treat['BSMTFINTYPE1'].fillna('NA')

mask = (df_null_treat_bsmt['BSMTCOND'].notnull()) & ((df_null_treat_bsmt['BSMTEXPOSURE'].isnull()) | (df_null_treat_bsmt['BSMTFINTYPE2'].isnull()))
filter = ~mask
df_null_treat = df_null_treat[filter].reset_index(drop=True)

df_null_treat['BSMTFINTYPE2'] = df_null_treat['BSMTCOND'].fillna('NA')
df_null_treat['BSMTEXPOSURE'] = df_null_treat['BSMTQUAL'].fillna('NA')

df_null_treat[bsmt_columns].isnull().sum()
```

```python
# NULLS PER COLUMN AND PERCENTAGE OF NULLS PER COLUMN

nulls_per_column = df_null_treat.isnull().sum()
percentage_nulls_per_column = df_null_treat.isnull().mean()
nulls_per_column = pd.DataFrame(nulls_per_column)
nulls_per_column = nulls_per_column.rename(columns={0: 'SUM'})
nulls_per_column['%'] = percentage_nulls_per_column.round(4)*100
nulls_per_column = nulls_per_column.sort_values(by='SUM', ascending=False)

nulls_per_column[nulls_per_column['SUM'] > 0]
```

```python
df_null_treat = df_null_treat.dropna(subset=['ELECTRICAL', 'MASVNRAREA'])
```

```python
# POOL ANALYSIS

df_null_treat_pool = df_null_treat[['POOLQC', 'POOLAREA']].copy()
pool_mask = (df_null_treat_pool['POOLQC'].notnull())
df_null_treat_pool = df_null_treat_pool[pool_mask]

df_null_treat_pool
```

```python
# POOL TREATMENT

df_null_treat['HASPOOL'] = np.where(df_null_treat['POOLQC'].isnull(), 0, 1).astype('int8')
df_null_treat = df_null_treat.drop(columns=['POOLQC', 'POOLAREA'])
```

```python
# FIREPLACE ANALYSIS

df_null_treat_fireplace = df_null_treat[['FIREPLACES', 'FIREPLACEQU']].copy()
fireplace_mask = (df_null_treat_fireplace['FIREPLACEQU'].isnull() & df_null_treat_fireplace['FIREPLACES'] > 0)
df_null_treat_fireplace = df_null_treat_fireplace[fireplace_mask]

df_null_treat_fireplace
```

```python
df_null_treat['FIREPLACEQU'] = df_null_treat['FIREPLACEQU'].fillna('NA')
```

```python
# GENERAL

df_null_treat['MISCFEATURE'] = df_null_treat['MISCFEATURE'].fillna('NA')
df_null_treat['ALLEY'] = df_null_treat['MISCFEATURE'].fillna('NA')
df_null_treat['FENCE'] = df_null_treat['MISCFEATURE'].fillna('NA')
df_null_treat['MASVNRTYPE'] = df_null_treat['MASVNRTYPE'].fillna('None')
df_null_treat['LOTFRONTAGE'] = df_null_treat['LOTFRONTAGE'].fillna(0)
```

```python
nulls_per_column = df_null_treat.isnull().sum()
percentage_nulls_per_column = df_null_treat.isnull().mean()

nulls_per_column = pd.DataFrame(nulls_per_column)
nulls_per_column = nulls_per_column.rename(columns={0: 'SUM'})
nulls_per_column['%'] = percentage_nulls_per_column.round(4)*100
nulls_per_column = nulls_per_column.sort_values(by='SUM', ascending=False)

nulls_per_column[nulls_per_column['SUM'] > 0]
```
