# Data Cleaning

```python
from pathlib import Path
import pandas as pd
import numpy as np
```

```python
# PATH DEFINITIONS
BASE_DIR = Path().resolve().parent

DATA_DIR = BASE_DIR / "data"
RAW_DATA_DIR = DATA_DIR / "data_raw"

CLEAN_DATA_DIR = DATA_DIR / "data_clean"
CLEAN_DATA_DIR.mkdir(exist_ok=True)
```

```python
df_raw = df_raw = pd.read_csv(RAW_DATA_DIR / "train.csv")
```

## String treatment

```python
df_str_treat = df_raw.copy()
```

```python
df_str_treat.columns = df_str_treat.columns.str.upper()
columns_str = df_str_treat.select_dtypes(include='string')

for col in columns_str.columns:
    df_str_treat[col] = df_str_treat[col].str.upper()

pd.set_option('display.max_columns', None)
df_str_treat.head()
```

## Null treatment

```python
df_null_treat = df_str_treat.copy()
```

```python
# CONFIRMATION IF THE COLUMNS THAT HAVE GARAGEYRBLT HAS NULL COLUMNS IN GARAGEQUAL, GARAGEFINISH, GARAGEFINISH AND GARAGECOND

garage_columns = ['GARAGEQUAL', 'GARAGEFINISH', 'GARAGETYPE', 'GARAGEYRBLT','GARAGECOND']

df_null_treat_garage = df_null_treat[garage_columns].copy()
df_null_treat_garage = df_null_treat_garage[df_null_treat_garage['GARAGEYRBLT'].isnull()].reset_index(drop='True')
df_null_treat_garage.isnull().sum()
```

```python
# GARAGEYRBLT 
df_null_treat['HASGARAGE'] = np.where(df_null_treat['GARAGEYRBLT'].isnull(), 0, 1).astype('int8')
df_null_treat = df_null_treat.drop(columns='GARAGEYRBLT')

garage_columns = ['GARAGEQUAL', 'GARAGEFINISH', 'GARAGETYPE', 'GARAGECOND', 'HASGARAGE']

# OTHERS
df_null_treat['GARAGEQUAL'] = df_null_treat['GARAGETYPE'].fillna('NA')
df_null_treat['GARAGEFINISH'] = df_null_treat['GARAGETYPE'].fillna('NA')
df_null_treat['GARAGETYPE'] = df_null_treat['GARAGETYPE'].fillna('NA')
df_null_treat['GARAGECOND'] = df_null_treat['GARAGETYPE'].fillna('NA')

df_null_treat[garage_columns].isnull().sum()
```

```python
basement_columns = ['BSMTFINTYPE2', 'BSMTEXPOSURE', 'BSMTCOND', 'BSMTQUAL', 'BSMTFINTYPE1']

df_null_treat_basement = df_null_treat[basement_columns].copy()
df_null_treat_basement.isnull().sum()
```

```python
# OTHERS
df_null_treat['BSMTCOND'] = df_null_treat['BSMTCOND'].fillna('NA')
df_null_treat['BSMTQUAL'] = df_null_treat['BSMTQUAL'].fillna('NA')
df_null_treat['BSMTFINTYPE1'] = df_null_treat['BSMTFINTYPE1'].fillna('NA')

mask = (df_null_treat_basement['BSMTCOND'].notnull()) & ((df_null_treat_basement['BSMTEXPOSURE'].isnull()) | (df_null_treat_basement['BSMTFINTYPE2'].isnull()))
filter = ~mask
df_null_treat = df_null_treat[filter].reset_index(drop=True)

df_null_treat['BSMTFINTYPE2'] = df_null_treat['BSMTCOND'].fillna('NA')
df_null_treat['BSMTEXPOSURE'] = df_null_treat['BSMTQUAL'].fillna('NA')

df_null_treat[basement_columns].isnull().sum()
```
